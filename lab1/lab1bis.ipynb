```python
def add_large_numbers(num1, num2):
    # Зробимо обидва числа однакової довжини, доповнюючи нулями зліва.
    max_len = max(len(num1), len(num2))
    num1 = num1.zfill(max_len)
    num2 = num2.zfill(max_len)
    
    result = []
    carry = 0
    
    # Розглядаємо числа зправа наліво.
    for i in range(max_len - 1, -1, -1):
        digit1 = int(num1[i])
        digit2 = int(num2[i])
        
        # Додаємо поточні цифри разом з попереднім переносом.
        total = digit1 + digit2 + carry
        
        # Визначаємо новий перенос.
        carry = total // 10
        
        # Додаємо поточну цифру до результату.
        result.append(str(total % 10))
    
    # Якщо залишився останній перенос, то додаємо його до результату.
    if carry:
        result.append(str(carry))
    
    # Перевертаємо результат, щоб отримати правильний порядок цифр.
    result.reverse()
    
    # Повертаємо результат як рядок.
    return ''.join(result)


def multiply_large_numbers(num1, num2):
    # Перетворюємо числа в зворотньому порядку, щоб легше множити зліва направо.
    num1 = num1[::-1]
    num2 = num2[::-1]

    # Ініціалізуємо список для збереження проміжних результатів.
    result = [0] * (len(num1) + len(num2))

    # Проводимо множення.
    for i in range(len(num1)):
        for j in range(len(num2)):
            result[i + j] += int(num1[i]) * int(num2[j])
            result[i + j + 1] += result[i + j] // 10
            result[i + j] %= 10

    # Позбавляємося від ведучих нулів.
    while len(result) > 1 and result[-1] == 0:
        result.pop()

    # Перетворюємо результат в рядок та перетворюємо його знову у правильний порядок.
    result = ''.join(map(str, result[::-1]))

    return result

def square_large_numbers(num):
  result = multiply_large_numbers(num, num)
  return result

def calculate_modulus_remainder(number, modulus):
    result = 0
    for digit in number:
        result = (result * 10 + int(digit)) % modulus
    return result

while True:
    print("-" * 40)
    print("Оберіть дію:")
    print("1. Додавання великих чисел")
    print("2. Множення великих чисел")
    print("3. Зведення числа в квадрат")
    print("4. Розрахунок залишку по модулю")
    print("5. Вийти з програми")
    choice = input("Введіть номер дії (1/2/3/4/5): ")

    if choice == '1':
        num1 = input("Введіть перше велике число: ")
        num2 = input("Введіть друге велике число: ")
        result = add_large_numbers(num1, num2)
        print("Результат додавання: ", result)
    elif choice == '2':
        num1 = input("Введіть перше велике число: ")
        num2 = input("Введіть друге велике число: ")
        result = multiply_large_numbers(num1, num2)
        print("Результат множення: ", result)
    elif choice == '3':
        num = input("Введіть велике число: ")
        result = square_large_numbers(num)
        print("Результат обчислення квадрату: ", result)
    elif choice == '4':
        num1 = input("Введіть велике число: ")
        num2 = int(input("Введіть модуль: "))
        result = calculate_modulus_remainder(num1, num2)
        print("Результат обчислення залишку по модулю: ", result)
    elif choice == '5':
        print("Програму завершено.")
        break
    else:
        print("Неправильний ввід, спробуйте ще.")
        continue

    another_calculation = input("Продовжити роботу? (y/n): ")
    if another_calculation.lower() != 'y':
        print("Програму завершено.")
        break


# Приклад використання:
# num1 = "1234567890123456789012345678901234567890"
# num2 = "9876543210987654321098765432109876543210"
# result = add_large_numbers(num1, num2)
# print(result)

# result = multiply_large_numbers(num1, num2)
# print(result)

# result = square_large_numbers(num1)
# print(result)
```
